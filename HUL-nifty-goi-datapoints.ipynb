```python
import pandas as pd
```

```python
df = pd.read_excel("HUL-master-excel.xlsx")
```

```python
df.head()
```

```python
df_daily = df.copy()
```

```python
df_daily.head()
```

```python
df_weekly = df.copy()
```

```python
df_weekly.head()
```

```python
df_monthly = df.copy()
df_monthly.head()
```

focus on `df_weekly`

```python
df_weekly['date'] = pd.to_datetime(df_weekly['date'])
```

```python
df_weekly.set_index('date', inplace=True)
```

```python
weekly_first = df_weekly.resample('W').first().dropna()
df_weekly = weekly_first
df_weekly.head()
```

```python
covid_end_date = '2023-05-11'
```

```python
df_weekly = df_weekly[df_weekly.index > covid_end_date]
```

```python
df_weekly.head()
```

```python
df_weekly.to_excel("weekly-data.xlsx", index=True)
```

focus on `df_monthly`

```python
df_monthly['date'] = pd.to_datetime(df_monthly['date'])
```

```python
df_monthly.set_index('date', inplace=True)
```

```python
monthly_first = df_monthly.resample('ME').first().dropna()
df_monthly = monthly_first
df_monthly.head()
```

```python
df_monthly = df_monthly[df_monthly.index > covid_end_date]
```

```python
df_monthly.head()
```

```python
df_monthly.to_excel("monthly-data.xlsx", index=True)
```

finally, focus on `df_daily`

```python
df_daily['date'] = pd.to_datetime(df_daily['date'])
```

```python
df_daily.set_index('date', inplace=True)
```
